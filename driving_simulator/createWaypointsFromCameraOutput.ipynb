# Create Waypoints from Camera Output
This Notebook shows how waypoints will be created from a file where the camera output is stored, when 'c' is pressed while someone is driving in the driving simulator

```python
# Read the file
f = open("strecke_7_ring_innen_fast.txt")
lines = f.read().split('\n')

idCounter = 1
speed = 130
vec_id = "strecke_7_ring_innen_fast"

xmlString = """<vehicle id="{id}">
	<modelPath>Models/Cars/drivingCars/Polizei/Car.j3o</modelPath>
	<mass>800</mass>
	<acceleration>6.0</acceleration>
	<decelerationBrake>8.7</decelerationBrake>
	<decelerationFreeWheel>2.0</decelerationFreeWheel>
	<engineOn>true</engineOn>
	<maxDistanceFromPath>5.0</maxDistanceFromPath>
	<curveTension>0.001</curveTension>
	<pathIsCycle>true</pathIsCycle>
	<pathIsVisible>true</pathIsVisible>
	<startWayPoint>WayPoint_1</startWayPoint>""".format(id=vec_id)

xmlString += "\t<wayPoints>"

for line in lines:
    # only check the line of the camera position
    if 'Camera Position' in line:
        start = line.find('(')
        end = line.find(')')
        values = line[start+1:end]
        values = values.replace(" ","").split(',')
        # now we have a list of 3 values (as strings)
        # We need to generate something like that:
        # <wayPoint id="WayPoint_1"> <translation><vector jtype="java_lang_Float" size="3"><entry>-520.5924</entry><entry>0.1</entry><entry>10.650283</entry></vector></translation><speed>90</speed></wayPoint>
        #print(values)
        
        xmlString += '\n\t\t<wayPoint id="WayPoint_{id}">'.format(id=idCounter)
        xmlString += '<translation><vector jtype="java_lang_Float" size="3">'
        xmlString += '<entry>{}</entry><entry>{}</entry><entry>{}</entry>'.format(values[0], values[1], values[2])
        xmlString += '</vector></translation>'
        xmlString += '<speed>{}</speed>'.format(speed)
        xmlString += '</wayPoint>'
        
        idCounter += 1

xmlString += "\n\t</wayPoints>\n</vehicle>"

print(xmlString)
```
